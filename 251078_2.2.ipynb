```python
import pandas as pd
```

```python
df = pd.read_csv('Dataset_2 - Sheet1.csv')
del df['Team']
df
```

```python
df['Team'] = ''
df.loc[df['Gender'] == 'F','Team'] = 'A'
df.loc[(df['Gender'] == 'M') & (df['Department'] == 'Finance'), 'Team'] = 'B'
df.loc[(df['Gender'] == 'M') & (df['Department'] == 'Software') & (df['Income'] > 40000), 'Team'] = 'A'
df.loc[(df['Gender'] == 'M') & (df['Department'] == 'Software') & (df['Income'] <= 40000), 'Team'] = 'B'
df.loc[(df['Gender'] == 'M') & (df['Department'] == 'HR') & (df['Income'] > 40000), 'Team'] = 'A'
df.loc[(df['Gender'] == 'M') & (df['Department'] == 'HR') & (df['Income'] <= 40000), 'Team'] = 'B'
df
```
