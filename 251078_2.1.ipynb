```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(pd.read_csv('Dataset_1 - Sheet1.csv'))
df
```

```python
X = df[['House Area', 'Number of rooms']].values
y = df['Price'].values.reshape(-1, 1)
X = np.c_[np.ones(X.shape[0]), X]
def gradient_descent(X, y, alpha=0.00000001, iterations=1000):
    m = len(y)
    theta = np.zeros((X.shape[1], 1))
    cost_history = []
    for i in range(iterations):
        predictions = X @ theta
        errors = predictions - y
        gradients = (1/m) * (X.T @ errors)
        theta = theta - alpha * gradients
        cost = (1/(2*m)) * np.sum(errors ** 2)
        cost_history.append(cost)
    return theta, cost_history
```

```python
theta_gd, cost_history = gradient_descent(X, y)
y_pred = X @ theta_gd
plt.figure(figsize=(8,5))
plt.scatter(y, y_pred)
plt.plot([y.min(), y.max()], [y.min(), y.max()])
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted Prices (Gradient Descent)")
plt.savefig(fname = 'Plot.png')
plt.show()

```
